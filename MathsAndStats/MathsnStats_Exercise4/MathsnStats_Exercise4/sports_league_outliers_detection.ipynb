## Problem Statement


You are a data scientist working for a sports analytics company. The dataset, named **"sports_league_data.csv"** includes a variety of attributes for players in the league. The columns in the dataset are as follows:

- **player_id:** A unique identifier for each player.
- **age:** Age of the player.
- **position:** The position the player typically plays in (e.g., Forward, Midfielder, Defender, Goalkeeper).
- **team:** The team the player belongs to.
- **games_played:** The total number of games played by the player in the season.
- **player_rating:** An overall performance rating for the player for the season.

**Import Necessary Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Task1

1. Import the data from the "sports_league_data.csv" file.
2. display the number of rows and columns. 
3. Display the first few rows of the dataset to get an overview.


```python
df = pd.read_csv("sports_league_data.csv")
print(df.shape)
df.head()
```

## Task2:

**Age Analysis:**
- Mean: Calculate the average age of players in the league.
- Standard Deviation Intervals:
    - Calculate the age range within one standard deviation of the mean.
    - Calculate the age range within two standard deviations of the mean.
    - Calculate the age range within three standard deviations of the mean.
  

```python
mean_age = round(df.age.mean(), 2)
std_age = round(df.age.std(), 2)
mean_age, std_age
```

```python
lower_age1 = round(mean_age - std_age, 2)
upper_age1 = round(mean_age + std_age, 2)
lower_age1, upper_age1
```

```python
lower_age2 = round(mean_age - 2*std_age, 2)
upper_age2 = round(mean_age + 2*std_age, 2)
lower_age2, upper_age2
```

```python
lower_age3 = round(mean_age - 3*std_age, 2)
upper_age3 = round(mean_age + 3*std_age, 2)
lower_age3, upper_age3
```

## Task3:

**Player Rating Analysis:**
- Mean: Determine the average player rating across the league.
- Standard Deviation Intervals:
    - Calculate the player rating range within one standard deviation of the mean.
    - Calculate the player rating range within two standard deviations of the mean.
    - Calculate the player rating range within three standard deviations of the mean.

```python
mean_player_rating = round(df.player_rating.mean(), 2)
std_player_rating = round(df.player_rating.std(), 2)
mean_player_rating, std_player_rating
```

```python
lower_player_rating1 = round(mean_player_rating - std_player_rating, 2)
upper_player_rating1 = round(mean_player_rating + std_player_rating, 2)
lower_player_rating1, upper_player_rating1
```

```python
lower_player_rating2 = round(mean_player_rating - 2*std_player_rating, 2)
upper_player_rating2 = round(mean_player_rating + 2*std_player_rating, 2)
lower_player_rating2, upper_player_rating2
```

```python
lower_player_rating3 = round(mean_player_rating - 3*std_player_rating, 2)
upper_player_rating3 = round(mean_player_rating + 3*std_player_rating, 2)
lower_player_rating3, upper_player_rating3
```

## Task4:

Create histogram with Kernel Density Estimation (KDE) plots for both the 'age' and 'player_rating' columns from the dataset. While constructing these plots, ensure to visually represent the standard deviation intervals on the graph.


- **'age'**

```python
sns.histplot(df.age, kde=True)
plt.axvline(mean_age, color='red', linestyle='--', label='Mean')

plt.axvline(mean_age - std_age, color='blue', linestyle='--', label='Mean - 1 Std Dev')
plt.axvline(mean_age + std_age, color='green', linestyle='--', label='Mean + 1 Std Dev')

plt.axvline(mean_age - 2*std_age, color='orange', linestyle='--', label='Mean - 2 Std Dev')
plt.axvline(mean_age + 2*std_age, color='purple', linestyle='--', label='Mean + 2 Std Dev')

plt.axvline(mean_age - 3*std_age, color='pink', linestyle='--', label='Mean - 3 Std Dev')
plt.axvline(mean_age + 3*std_age, color='brown', linestyle='--', label='Mean + 3 Std Dev')

plt.title('Histogram with Kernel Density Estimation (KDE) plot for the "age"')
plt.legend()
plt.show()
```

- **'player_rating'**

```python
sns.histplot(df.player_rating, kde=True)
plt.axvline(mean_player_rating, color='red', linestyle='--', label='Mean')

plt.axvline(mean_player_rating - std_player_rating, color='blue', linestyle='--', label='Mean - 1 Std Dev')
plt.axvline(mean_player_rating + std_player_rating, color='green', linestyle='--', label='Mean + 1 Std Dev')

plt.axvline(mean_player_rating - 2*std_player_rating, color='orange', linestyle='--', label='Mean - 2 Std Dev')
plt.axvline(mean_player_rating + 2*std_player_rating, color='purple', linestyle='--', label='Mean + 2 Std Dev')

plt.axvline(mean_player_rating - 3*std_player_rating, color='pink', linestyle='--', label='Mean - 3 Std Dev')
plt.axvline(mean_player_rating + 3*std_player_rating, color='brown', linestyle='--', label='Mean + 3 Std Dev')

plt.title('Histogram with Kernel Density Estimation (KDE) plot for the "player_rating"')
plt.legend()
plt.show()
```
