We have movies dataset. We want to do some basic analysis such as,

1. What is the min, max and average rating for all movies?
1. What is the min, max and average rating for all bollywood movies?
1. What is the min, max and average rating for all hollywood movies?

We will see how this can be done first without using Pandas and then using Pandas

### Option 1 - Movies analysis without using Pandas

```python
import csv

def calculate_rating_stats(data, industry=None):
    ratings = []
    for row in data:
        if row[3]!='NULL' and (not industry or row[1]==industry):
            ratings.append(float(row[3]))
    max_rating = max(ratings)
    min_rating = min(ratings)
    avg_rating = sum(ratings) / len(ratings)
    return max_rating, min_rating, avg_rating

with open('movies.csv') as f:
    data = list(csv.reader(f))
    header = data[0]
    data = data[1:]
    
    max_rating, min_rating, avg_rating = calculate_rating_stats(data)
    print(f"All records: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
    
    max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Bollywood")
    print(f"Bollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
    
    max_rating, min_rating, avg_rating = calculate_rating_stats(data, industry="Hollywood")
    print(f"Hollywood: Min rating = {min_rating}, Max rating = {max_rating}, Avg rating = {avg_rating}")
```

### Option 2 - Movies analysis using Pandas

**Print highest, lowest and average movie rating**

```python
import pandas as pd

df = pd.read_csv("movies.csv")
df
```

```python
df.head()
```

```python
print(f'Max rating is {df.imdb_rating.max()}')
```

```python
print(f'Min rating is {df.imdb_rating.min()}')
```

```python
print(f'Average rating is {df.imdb_rating.mean()}')
```

**Print highest, lowest and average movie rating in bollywood and hollywood individually**|

```python
df_hollywood = df.imdb_rating[df.industry == 'Hollywood']
df_bollywood = df.imdb_rating[df.industry == 'Bollywood']

print(f'Highest rating of Hollywood movies is {df_hollywood.max()}')
print(f'Highest rating of Bollywood movies is {df_bollywood.max()}')
```

```python
print(f'Lowest rating of Hollywood movies is {df_hollywood.min()}')
print(f'Lowest rating of Bollywood movies is {df_bollywood.min()}')
```

```python
print(f'Average rating of Hollywood movies is {df_hollywood.mean()}')
print(f'Average rating of Bollywood movies is {df_bollywood.mean()}')
```

Complete code of pandas analysis

```python
df = pd.read_csv("movies.csv")
print("All movies: Min=", df.imdb_rating.min(), "Max=", df.imdb_rating.max(), "Avg=", df.imdb_rating.mean())

df_b = df[df.industry=="Bollywood"]
df_h = df[df.industry=="Hollywood"]

print("Bollywood: Min=", df_b.imdb_rating.min(), "Max=", df_b.imdb_rating.max(), "Avg=", df_b.imdb_rating.mean())
print("Hollywood: Min=", df_h.imdb_rating.min(), "Max=", df_h.imdb_rating.max(), "Avg=", df_h.imdb_rating.mean())
```
